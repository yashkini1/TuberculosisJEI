```python
# Import necessary packages
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path2018to2022 = r'C:\Users\kiniy\Python Class\Test File_2018 to 2022 - Test File_2018 to 2022 (3).csv'
path2013to2017 = r'C:\Users\kiniy\Python Class\Test File_2013 to 2017 - Test File_2013 to 2017 (5).csv'
```

```python
#FIGURE 5 (MODEL 1): FEATURE IMPORTANCE FROM 2018 TO 2022
# Load the dataset
df = pd.read_csv(path2018to2022)
```

```python
# Define the columns to be used in the analysis
categorical_columns = ['race', 'age_group', 'sex', 'not_us_born', 'hiv_positive']
target_column = 'incidence_rate'
```

```python
# Select the relevant columns
df_relevant = df[categorical_columns + [target_column]]
```

```python
# Encode the categorical feature variables
encoder = OneHotEncoder(sparse_output=False)
X_encoded = pd.DataFrame(encoder.fit_transform(df_relevant[categorical_columns]))
X_encoded.columns = encoder.get_feature_names_out(categorical_columns)
```

```python
# Encode the target variable
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(df_relevant[target_column])
```

```python
# Combine the encoded features with the target column for correlation analysis
df_encoded = pd.concat([X_encoded, pd.Series(y_encoded, name=target_column)], axis=1)
```

```python
# Calculate correlation matrix
correlation_matrix = df_encoded.corr()
```

```python
# Extract correlation of each feature with the target variable
feature_importance = correlation_matrix[target_column].drop(target_column)
```

```python
feature_importance_df = pd.DataFrame({
    'Feature': feature_importance.index,
    'Importance': feature_importance.values
})
```

```python
# Sort the features by importance
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)
```

```python
# Plot the feature importance
plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=feature_importance_df)
plt.title('Feature Importance in Dataset (2018 to 2022)')
plt.show()
```
